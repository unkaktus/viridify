```python
import numpy as np

import matplotlib.pyplot as plt

plt.rc('figure', dpi=60)
```

```python
data = np.loadtxt('grid_pBHNS_SLy144_0159_rcut100.txt', skiprows=1)
data = data[:,3].reshape((100, 100, 100))
data = data[49,:,:]
data = np.log10(data).T
```

```python
cmaps = ['jet', 'magma', 'viridis']
for cmap in cmaps:
    im = plt.imshow(data, cmap=cmap, vmin=-17)
    cbar = plt.colorbar(im)
    cbar.set_ticklabels([])
    plt.yticks([])
    plt.xticks([])
    plt.savefig(f'nsbh_{cmap}.png', bbox_inches='tight')
    plt.figure().clear()
```
